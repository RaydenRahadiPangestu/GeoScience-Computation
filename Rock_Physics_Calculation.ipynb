```python
K_fluid_g =  1/ ((sat_wat/Kw) + (((1-sat_wat)/kg)))   # gass
df['K_sat gas'] = kdry + (1 - kdry/Km)**2 / ((df['Correct Porosity'] / K_fluid_g) + ((1-df['Correct Porosity'])/Km) - (kdry/Km**2))

K_fluid_o =  1/ ((sat_wat/Kw) + (((1-sat_wat)/ko)))   # oil
df['K_sat oil'] = kdry + (1 - kdry/Km)**2 / ((df['Correct Porosity'] / K_fluid_o) + ((1-df['Correct Porosity'])/Km) - (kdry/Km**2))
```
